```python
import matplotlib.pyplot as plt
import numpy as np

# Set up a 3x3 grid
plt.figure(figsize=(12, 8))

# First graph
xpoints = np.array([1, 8])
ypoints = np.array([3, 10])
plt.subplot(3, 3, 1)
plt.plot(xpoints, ypoints, 'o')
plt.title("Title 1")
plt.xlabel("blue colour")
plt.ylabel("blue colour")

# Second graph
x = np.array([2, 5, 2, 5])
y = np.array([6, 2, 6, 2])
plt.subplot(3, 3, 2)
plt.plot(y, marker="*", color='orange', linewidth=2.5)
plt.plot(x, linestyle='dashed', linewidth=2.5)


# Third graph
ypoints = np.array([1, 5, 3, 7])
plt.subplot(3, 3, 3)
plt.plot(ypoints, marker='o', ms=20, mfc='yellow', mec='black')


# Fourth graph
x = np.array([80, 85, 90, 95, 100, 105, 110, 115, 120, 125])
y = np.array([240, 250, 260, 270, 280, 290, 300, 310, 320, 330])
plt.subplot(3, 3, 4)
plt.plot(x, y)
plt.grid()
plt.title("Grid")

# Fifth graph
x = np.array([5, 7, 8, 7, 2, 17, 2, 9, 4, 11, 12, 9, 6])
y = np.array([99, 86, 87, 88, 111, 86, 103, 87, 94, 78, 77, 85, 86])
sizes = np.array([20, 50, 100, 200, 500, 1000, 60, 90, 10, 300, 600, 800, 75])
colors = np.array([0, 10, 20, 30, 40, 45, 50, 55, 60, 70, 80, 90, 100])
plt.subplot(3, 3, 5)
plt.scatter(x, y, s=sizes, c=colors, alpha=0.5)


# Sixth graph
x = np.array(["A", "B", "C", "D"])
y = np.array([3, 8, 1, 10])
plt.subplot(3, 3, 6)
plt.bar(x, y, width=0.1, color='red')


# Seventh graph
y = np.array([35, 20, 20, 25])
mylabels = ["Apples", "Bananas", "Cherries", "Dates"]
myexplode = [0.2, 0, 0, 0]
mycolors = ["green", "hotpink", "b", "yellow"]
plt.subplot(3, 3, 7)
plt.pie(y, startangle=90, explode=myexplode, shadow=True, colors=mycolors, labels=mylabels)
plt.legend(loc='best')


# Eighth graph
x = np.array([80, 85, 90, 95, 100, 105, 110, 115, 120, 125])
y = np.array([240, 250, 260, 270, 280, 290, 300, 310, 320, 330])
plt.subplot(3, 3, 8)
plt.plot(x, y)
plt.grid(axis='y')

# Ninth graph
x = np.random.normal(170, 10, 250)
plt.subplot(3, 3, 9)
plt.hist(x)


# Adjust layout and show plot
plt.tight_layout()
plt.suptitle("Graph", fontsize=16)
plt.subplots_adjust(top=0.9)  # Adjust top to make space for the suptitle
plt.show()
```
